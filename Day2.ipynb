# Problem 1: Unique Words in a Sentence


```python
def unique_words(sentence: str):
#    Returns unique words and their count from a sentence.
 
    words = sentence.lower().split()   # Convert to lowercase for uniformity
    unique = set(words)                # Remove duplicates
    return unique, len(unique)


# Example usage
sentence = "python is easy and python is powerful"
unique, count = unique_words(sentence)

print("Unique words count:", count)
print("Unique words:", unique)
```

# Problem 2: Highest Salary from Employee Data

```python
def highest_salary(employees: dict):
   
  #  Returns employee with highest salary.
  
    name = max(employees, key=employees.get)
    return name, employees[name]


employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

name, salary = highest_salary(employees)

print(f"Highest Salary: {name} - {salary}")
```

# Problem 3: Find Maximum and Minimum Values


```python
# Problem 3: Find Max and Min

def find_max_min(numbers: list):
    """
    Returns maximum and minimum values from list.
    """
    maximum = minumum = numbers[0]

    for num in numbers:
        if num > maximum:
            maximum = num
        if num < minumum:
            minumum = num

    return maximum, minumum


numbers = [45, 22, 89, 10, 66]

max_val, min_val = find_max_min(numbers)

print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

# Problem 4: Count Products Above Price Threshold

```python
def count_above_threshold(prices: list, threshold: int = 1000):
#    Counts number of products above given threshold.
 
    count = sum(1 for price in prices if price > threshold)
    return count


prices = [450, 1200, 899, 1500, 300]

result = count_above_threshold(prices)

print("Products above 1000:", result)
```

# Problem 5: Calculate Attendance Percentage

```python
def calculate_attendance(attendance: list):
 
 #   Calculates attendance percentage.
  
    total_days = len(attendance)
    present_days = attendance.count("P")
    percentage = (present_days / total_days) * 100
    return round(percentage, 2)


attendance = ["P", "P", "A", "P", "P"]

result = calculate_attendance(attendance)

print("Attendance Percentage:", result)
```

# Problem 7: Count Character Frequency

```python
def character_frequency(text: str):

# Counts frequency of each character.
 
    frequency = {}

    for char in text:
        frequency[char] = frequency.get(char, 0) + 1

    return frequency


text = "python"

result = character_frequency(text)

print(result)
```

# Problem 9: Check if Key Exists in Dictionary

```python
def check_key(dictionary: dict, key):
    
#   Checks if key exists in dictionary.
    if key in dictionary:
        return "Employee exists"
    else:
        return "Employee does not exist"


employees = {"Ravi": 75000, "Anita": 68000}

print(check_key(employees, "Ravi"))
```

# Problem 10: Calculate Average Marks

```python
def calculate_average(marks: list):
    
# Calculates average of marks.

    return round(sum(marks) / len(marks), 2)


marks = [70, 75, 80, 65]

result = calculate_average(marks)

print("Average Marks:", result)
```
