# Gaming leaderboard System


# should include
1. Accept scores from multiple matches
2. calculate total score
3. determine player rank
4. flag professional player

# Rules- 
- Total>= 1000 -> Pro Player
# Rank assignment
-  Grater then equal to 1200 -> Diamond
-  Grater then equal to 900 -> Gold
-  Grater then equal to 600 -> Silver
- else -> Bronze

```python
def analyze_player(scores):
    total=0
    for score in scores:
        total = total+score
    # Determine ranks
    if total>= 1200:
        rank= "Diamond"
    elif total>= 900:
        rank= "Gold"
    elif total>=600:
        rank="Sliver"
    else:
        rank="Bronze"
   # Professional Tag
    if total>=1000:
        status= "Pro player"
    else:
        status= "Normal Player"
    return total,rank,status
# user input
scores=[]
matches= int(input("enter number of matches played:"))
for i in range(matches):
    s=int(input(f"Enter score for match{i+1}:"))
    scores.append(s)

# function call
total,rank,status= analyze_player(scores)
# output
print("Total score=",total)
print("Rank=",rank)
print("Status=",status)
    
    
```

# Problem 1: Smart Parking Lot Management System

```python
def manage_parking(capacity, vehicle_logs):
    current_parked = 0
    peak_usage = 0

    for log in vehicle_logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT" and current_parked > 0:
            current_parked -= 1

        # Track peak usage
        if current_parked > peak_usage:
            peak_usage = current_parked

    print("Currently Parked Vehicles:", current_parked)

    if current_parked > capacity:
        print("Parking Status: Capacity Exceeded")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak_usage)


# Sample Input
manage_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

# Problem 2: Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):
    # Base time calculation (5 minutes per km)
    eta = distance * 5

    # Traffic adjustment
    if traffic == "High":
        eta += 15
    elif traffic == "Medium":
        eta += 8

    # Weather adjustment
    if weather == "Rainy":
        eta += 10
    elif weather == "Storm":
        eta += 20

    print("Estimated Delivery Time:", eta, "minutes")


# Sample Input
estimate_delivery_time(8, "High", "Rainy")
```

# Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
def analyze_occupancy(total_seats, booked_seats):
    booked_count = 0

    for seat in booked_seats:
        booked_count += seat

    occupancy_percentage = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy_percentage), "%")

    if occupancy_percentage == 100:
        print("Show Status: Housefull")
    elif occupancy_percentage >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Sample Input (150 booked seats out of 200)
analyze_occupancy(200, [1]*150)
```

# Problem 4: Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    total = 0

    for reading in cpu_readings:
        total += reading

    average_cpu = total / len(cpu_readings)

    print("Average CPU Load:", int(average_cpu), "%")

    if average_cpu < 50:
        print("Server Status: Normal")
    elif average_cpu <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
classify_server_load([45, 60, 70, 85, 90])
```

# Problem 5: Smart Classroom Resource Usage Monitor

```python
def monitor_resources(resource_usage):
    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:  # Overuse threshold
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(resources)
```

# Problem 6: Online Event Registration Capacity Controller

```python
def manage_registrations(capacity, registrations):
    if registrations <= capacity:
        print("Confirmed Registrations:", registrations)
        print("Waitlisted Users:", 0)
        print("Registration Status: Open")
    else:
        waitlist = registrations - capacity
        print("Confirmed Registrations:", capacity)
        print("Waitlisted Users:", waitlist)
        print("Registration Status: Closed")


# Sample Input
manage_registrations(100, 105)
```
