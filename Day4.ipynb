# Hotel Booking system
- customer age must be 18 or above
-  stay duration must 1 to 10 days
-  Room type - standard or Deluxe
-  base price - 3000
-  delux room 2000 extra per day

# Create a function 
- Validate booking rules
- calculate total bill
- return booking status

```python
def  calculate_hotel_bill(age,days,room_type):
    base_price= 3000
    if age<18:
        return "Booking Rejected: Age must be 18 or above "
    if days<=0 or days>10:
        return "Booking Rejected: Invalid days entered"
    if room_type not in ["standard", "deluxe"]:
        return "Booking Rejected: Invalid room type"
    total = base_price* days
    if room_type == "deluxe":
        total=total+2000*days
    return f"Booking Successful! Total bill is :{total}"

#user input
age= int(input("enter age:"))
days= int(input("enter days:"))
room_type=input("enter room type(standard/deluxe):").lower()

#Function call
result = calculate_hotel_bill(age,days,room_type)
print(result)
```

# Problem 1: Mobile Recharge Validation System

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    elif amount not in valid_plans:
        print("Invalid plan selected.")
        return False
    else:
        print("Recharge successful!")
        return True


# Retry using while loop
while True:
    try:
        amount = int(input("Enter recharge amount: "))
        if validate_recharge(amount):
            break
    except ValueError:
        print("Please enter a valid number.")
```

# Problem 2: Inventory Reorder Alert System

```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product} → Reorder Alert (Stock: {stock})")
        else:
            print(f"{product} → Stock OK (Stock: {stock})")


inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18
}

check_inventory(inventory)
```

# Problem 3: Student Result Processing System

```python
def process_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


marks = [60, 55, 70, 40]
result = process_result(marks)
print("Result:", result)
```

# Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak_hour):
    base_fare = 50
    fare = base_fare + (distance * 12)

    if peak_hour:
        fare += fare * 0.25  # 25% extra

    return fare


# Retry using while loop
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak_input = input("Is it peak hour? (yes/no): ").lower()

        peak_hour = True if peak_input == "yes" else False

        total_fare = calculate_fare(distance, peak_hour)
        print("Total Fare: ₹", total_fare)
        break

    except ValueError:
        print("Invalid input. Please try again.")
```

# Problem 5: Employee Attendance Eligibility Checker

```python
def check_attendance(attendance_list):
    present_days = 0

    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
status = check_attendance(attendance)
print("Status:", status)
```

# Problem 6: Password Strength Checker

```python
def check_password(password):
    special_characters = "@#$"

    if len(password) < 8:
        return "Weak Password (Minimum length 8 required)"

    has_digit = False
    has_special = False

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_characters:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must include digit and special character)"


password = input("Enter password: ")
print(check_password(password))
```
