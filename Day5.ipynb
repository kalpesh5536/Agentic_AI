# HealthCare
- P.S -> Healthcare monitoring system
- Based on the patients heart rate , classify health status
- Real Time use- smart watches/ fitness tracker.
- ICU patient monitorig system
- emergency alert system

# Classfication Rules:
- when patient is at rest, heart rate
- below 60 -> low
- 60 to 100 -> normal
- above 100 -> High

- When patient is at active state
- below 60 -> low
-  60 to 120 -> normal
-  above 120 -> high
# Risk alert rules: 
- if status is low or high -> medical attention required
- if status is normal-> stable condition

```python
def classify_heart_rate(heart_rate, activity_status):
    # check if patient is at rest
    if activity_status =="rest":
        if heart_rate<60:
            status=" Your heart rate is low"
        elif heart_rate<=100:
            status= "Your heart rate is normal"
        else:
            status= "Your heart rate is high"
    # check if patient is at active
    elif activity_status=="active":
        if heart_rate<60:
            status= "Your heart rate is low"
        elif heart_rate<=120:
            status= "Your heart rate is normal"
        else:
            status= "Your heart rate is high"
    else:
        return "Invalid activity status"
    # Risk alert rules
    if "normal" in status:
        risk=  "You are in stable condition"
    else:
        risk= "You need medical attention ASAP"
    # return final result
    return status, risk
# user inputs
heart_rate= int(input("enter patient's heart rate"))
activity_status =input("enter activity status(rest/active):").lower()

# function call

result=classify_heart_rate(heart_rate, activity_status)
print(result)
```

# Agriculture - Crop yield analysis
- P.S - farmers enter daily crop yields value(KG)
1. collect yield data from user
2. calculate total yield and average yield
3. decide whether crop production good or poor

# Rules
- Average yield >= 50 -> good
- else poor production

```python
def crop_yield_analysis(yields):
    total=0
# calculate total yield
    for value in yields:
        total=total+value

# average
    average= total/len(yields)

# decide yield status
    if average>=50:
        status= " it is Good Crop production"
    else: 
        status= "it is Bad crop production"
# return all results
    return total,average,status

# empty list
daily_yields= []
# ask days
days= int(input("enter the number of days:"))
for i in range(days):
    yield_value=int(input(f"Enter yield for day {i+1}(kg):"))
    daily_yields.append(yield_value)

# call function
total,average,result= crop_yield_analysis(daily_yields)

# output
print("Daily yields:",daily_yields)
print("Total",total)
print("average",average)
print("status", result)
```

# Problem 3: Rainfall Adequacy Checker

```python
def analyze_rainfall(rainfall_data, required_level):
    total = 0

    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    print("Average Rainfall:", average)

    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


rainfall = [70, 75, 68, 80, 67]
analyze_rainfall(rainfall, 70)
```

# Problem 4: Duplicate Account Detection

```python
def check_duplicates(usernames):
    unique_users = set(usernames)

    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


users = ["user1", "user2", "user1", "user3"]
check_duplicates(users)
```

# Problem 9: Attendance Eligibility System

```python
def check_exam_eligibility(attendance_list):
    present_days = 0

    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


attendance = ["P", "P", "A", "P", "P"]
check_exam_eligibility(attendance)
```

# Problem 10: Smart Electricity Bill Analyzer

```python
def calculate_electricity_bill(units):
    bill = 0

    # Billing Slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Status:", usage)


calculate_electricity_bill(250)
```
