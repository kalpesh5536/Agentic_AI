#  Education : Student Profile System
- P.S--> I want to create student class to:
1. store student marks and name
2. calculate average
3. display result(pass/fail)

- oops concept- class, object, method calling

```python
class Students:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
    def calculate_average(self):
        return sum(self.marks)/len(self.marks)
    def show_result(self):
        avg=self.calculate_average()
        if avg>=50:
            print(self.name,"Pass")
        else:
            print(self.name,"Fail")
    # Object creation
student= Students("kalpesh",[80,60,60])
student.show_result()

s= Students("sarika",[10,20,30])
s.show_result()
```

# Banking: Bank Account system

- P.S--> Create bank account class:
1. stores a/c holder name and balance
2. allow deposit
3. allow withdraw
4. display balance

```python
class BankAccount:
    def __init__(self,name,balance):
        self.name=name
        self.balance=balance
        
    def deposit(self,amount):
        self.balance=self.balance+amount
        print("deposited:",amount)
    
    def withdraw(self,amount):
        if amount<=self.balance:
            self.balance -=amount
            print("Withdraw amount is=",amount)
        else:
            print("Insufficent balance")
    def show_balance(self):
        print("current balance in account is",self.balance)
# create object

account= BankAccount("Kalpesh",5000)
account.deposit(2000)
account.withdraw(3000)
account.show_balance()
    
```

# Transportation : vehicle infor system
- vehicle class that stores:
1. vehicle number, fuel level(liters)
2. ability to refuel and display status(fuel level)
3. print vehicle number, fuel refuel

```python
class vehicle:
    def __init__(self, vehicle_number, fuel_lit):
        self.vehicle_number = vehicle_number
        self.fuel_lit = fuel_lit

    def refuel(self, lit):
        self.fuel_lit += lit
        print(f"{lit} liters added to vehicle {self.vehicle_number}")

    def display_status(self):
        print(f"Vehicle Number: {self.vehicle_number}")
        print(f"Current Fuel Level: {self.fuel_lit} liters")

        if self.fuel_lit < 10:
            print("Low fuel alert")
        else:
            print("Fuel sufficient")


# Create object
v = vehicle("MH123213", 8)

v.display_status()
v.refuel(15)
v.display_status()
```

# Problem 1: Smart Light Controller

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"  # Default state

    def turn_on(self):
        self.status = "ON"
        print(f"{self.name} is {self.status}")

    def turn_off(self):
        self.status = "OFF"
        print(f"{self.name} is {self.status}")


# Sample Input
light = SmartLight("Bedroom Light")
light.turn_on()
```

# Problem 2: Employee ID Card System

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Sample Input
employee = Employee("Rahul", "EMP102", "IT")
employee.display_id_card()
```

# Problem 3: Mobile Contact Record

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def save_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Sample Input
contact = Contact("Anita", 9876543210)
contact.save_contact()
```

# Problem 4: Product Price Tag Generator

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


# Sample Input
product = Product("Headphones", 2499)
product.print_price_tag()
```

# Problem 5: Movie Rating Display System

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


# Sample Input
movie = Movie("Inception", 4.8)
movie.display_movie()
```

# Problem 6: Delivery Address Manager

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def print_delivery_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Sample Input
delivery = Delivery("Suman", "Hyderabad")
delivery.print_delivery_details()
```
